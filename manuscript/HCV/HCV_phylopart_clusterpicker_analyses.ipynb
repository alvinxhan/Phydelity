# Running PhyloPart and ClusterPicker on HCV France dataset

This jupyter notebook performs clustering analyses using PhyloPart and ClustePicker and compares the results with Phydelity output clusters. Note that there are both Python and R codes in this notebook. 

## Import Python modules and libraries

```python
from __future__ import division
from sklearn.metrics import silhouette_score
from matplotlib import pyplot as plt
import pandas as pd 
import numpy as np
import csv 
import ete3
import re
import os
import subprocess
import itertools 
import random 
import string
import platform
if platform.system() == "Darwin": # mac home
    home = "Users"
else: 
    home = "home"
```

## Parameters 

Note that a range of genetic distance thresholds will be evaluated - ClusterPicker: 0.005-0.1 substitutions/site; PhyloPart: 0.5-10 percentile of global patristic distance distribution. 

```python
thresholds = np.arange(0.005, 0.101, 0.001) # genetic distance threshold range 
genotypes = ["1a", "4d"]

# path to jarfile for PhyloPart and ClusterPicker --> CHANGE THIS AS NECESSARY 
phylopart_jarfile = "/{}/alvin/Dropbox/softwares/PhyloPart_v2.1/PhyloPart_v2.1.jar".format(home)
clusterpicker_jarfile = "/{}/alvin/Dropbox/softwares/ClusterPicker_1.2.3.jar".format(home)
```

## Run PhyloPart and ClusterPicker

```python
phylopart_df = {"cluster":[], "leaf":[], "threshold":[], "genotype":[]}
clusterpicker_df = {"cluster":[], "leaf":[], "threshold":[], "genotype":[]}

for genotype in genotypes:
    
    ori_treefile = "RAxML_bestTree.FR_MSM_HCV_{}".format(genotype)
    
    tree = ete3.Tree(ori_treefile)
    # get midpoint outgroup 
    outgroup = tree.get_midpoint_outgroup()
    tree.set_outgroup(outgroup)
    tree.ladderize()

    # write midpoint rooted tree in newick format
    treefile = "{}_midpoint_rooted.newick".format(ori_treefile)
    tree.write(format=5,outfile=treefile)

    fastafile = "mafft_FR_MSM_HCV_{}_nuc.fasta".format(genotype) # FASTA alignment file
    
    for t in thresholds:
        
        # Run PhyloPart 
        outfname = "ppOUT.csv" # output filename 

        # run PhyloPart with current threshold 
        cmd = ["java", "-Xss2m", "-jar", phylopart_jarfile, treefile, str(t), "-b0", "-o{}".format(outfname)]
        subprocess.call(cmd)

        # read output file 
        with open(outfname, "rb") as fhandle:
            for l, line in enumerate(csv.reader(fhandle)):
                if l == 0:
                    continue 

                cluster, leaf = int(line[0]), re.sub("_{}$".format(genotype), "", line[2])
                if cluster == 0:
                    continue 

                # save cluster, leafname and threshold into dataframe
                phylopart_df["cluster"].append(cluster)
                phylopart_df["leaf"].append(leaf)
                phylopart_df["threshold"].append(t)
                phylopart_df["genotype"].append(genotype)
        
        # Run ClusterPicker
        outfname = "{}_clusterPicks_log.txt".format(re.sub("\.[^\.]+$", "", ori_treefile))
        cmd = ["java", "-Xss2m", "-jar", clusterpicker_jarfile, fastafile, ori_treefile, "0", "0", str(t)]
        subprocess.call(cmd)

        # read output file 
        with open(outfname, "rb") as fhandle:
            for l, line in enumerate(reversed(list(csv.reader(fhandle, delimiter="\t")))):
                if l == 0:
                    continue 
                if line[0] == "ClusterNumber":
                    break 

                cluster, clustered_leaves = line[0], line[3]
                for leaf in re.sub("(\[|\])", "", clustered_leaves).split(", "):
                    leaf = re.sub("_{}$".format(genotype), "", leaf)
                    clusterpicker_df["cluster"].append(int(cluster))
                    clusterpicker_df["leaf"].append(leaf)
                    clusterpicker_df["threshold"].append(t)
                    clusterpicker_df["genotype"].append(genotype)

phylopart_df = pd.DataFrame(phylopart_df)
clusterpicker_df = pd.DataFrame(clusterpicker_df)

phylopart_df.head()
```

```python
# get silhouette index (SI)
def get_SI(df, t):
    # filter dataframe based on threshold 
    fildf = df[np.isclose(df["threshold"], t)]

    if len(fildf) == 0:
        return 0.

    # get set of unique clusters 
    cluster_set = list(set(fildf.cluster))

    labels_pred = [] # cluster labels 
    for leaf in all_leaf_names: 
        if leaf in list(fildf.leaf):
            # for leaves that were clustered, append its cluster label to labels_pred
            cluster = fildf[fildf.leaf == leaf]["cluster"].iloc[0]
        else: 
            # for unclustered leaves, they will just have a unique cluster label each 
            cluster = random.choice(list(set(range(1000)) - set(cluster_set)))
            cluster_set.append(cluster)

        labels_pred.append(cluster)

    if len(set(labels_pred)) == 1:
        return np.nan 

    return silhouette_score(X, labels_pred)

genotype_to_optimalSI_threshold = {}
for genotype in genotypes:
    
    treefile = "RAxML_bestTree.FR_MSM_HCV_{}".format(genotype)
    
    # read tree file using ete3
    tree = ete3.Tree(treefile)
    leaves = [_ for _ in tree.get_leaves()] # all leaf nodes except outgroup
    all_leaf_names = [re.sub("_{}$".format(genotype), "", leaf.name) for leaf in leaves] # all leaf names 
    N = len(all_leaf_names)

    # get array of inter-leaf distances 
    X = np.zeros((N, N))
    for i, j in itertools.combinations(range(N), 2):
        X[(i,j)] = X[(j,i)] = leaves[i].get_distance(leaves[j])

    # for each threshold
    pp_SI = np.zeros(len(thresholds)) # phylopart SI
    cp_SI = np.zeros(len(thresholds)) # clusterpicker SI

    for _t, t in enumerate(thresholds): 
        # First, we analyse results from PhyloPart 
        pp_SI[_t] = get_SI(phylopart_df[phylopart_df.genotype == genotype], t)

        # Repeat for ClusterPicker 
        cp_SI[_t] = get_SI(clusterpicker_df[clusterpicker_df.genotype == genotype], t)

    # plot t vs SI 
    plt.plot(thresholds, pp_SI, "#ef8a62", label="PhyloPart")
    plt.plot(thresholds, cp_SI, "#67a9cf", label="ClusterPicker")
    plt.yticks(np.arange(0, 0.51, 0.05))
    
    # annotate SI max 
    pp_ymax = np.max(pp_SI[~np.isnan(pp_SI)])
    pp_xmax = thresholds[list(pp_SI).index(pp_ymax)]
    plt.annotate(r'$t_{{optimal}}$'.format(pp_xmax, pp_ymax), xy=(pp_xmax, pp_ymax), xytext=(pp_xmax+0.007, pp_ymax+0.05), 
                 arrowprops=dict(facecolor='#ef8a62', width=1.5, headwidth=12, headlength=8, shrink=0.05),)

    cp_ymax = np.max(cp_SI[~np.isnan(cp_SI)])
    cp_xmax = thresholds[list(cp_SI).index(cp_ymax)]
    plt.annotate(r'$t_{{optimal}}$'.format(cp_xmax, cp_ymax), xy=(cp_xmax, cp_ymax), xytext=(cp_xmax+0.007, cp_ymax+0.05), 
                 arrowprops=dict(facecolor='#67a9cf', width=1.5, headwidth=12, headlength=8, shrink=0.05),)
    plt.ylabel("$SI$")
    plt.xlabel("Threshold ($t$)")
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.savefig("./SI_HCV_{}.pdf".format(home, genotype),bbox_inches="tight")
    plt.show()
    
    best_pp_t = np.min([t for _t, t in enumerate(thresholds) if pp_SI[_t] == np.amax(pp_SI)][0])
    try:
        genotype_to_optimalSI_threshold[genotype]["pp"] = best_pp_t
    except:
        genotype_to_optimalSI_threshold[genotype] = {"pp":best_pp_t}
    
    best_cp_t = np.min([t for _t, t in enumerate(thresholds) if cp_SI[_t] == np.amax(cp_SI[~np.isnan(cp_SI)])])
    genotype_to_optimalSI_threshold[genotype]["cp"] = best_cp_t
    
genotype_to_optimalSI_threshold = pd.DataFrame(genotype_to_optimalSI_threshold)
genotype_to_optimalSI_threshold
```
